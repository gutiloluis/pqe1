```python
X = np.linspace(0,8,100)
pdf = sps.norm.pdf(X, loc=3, scale=0.5)
cdf = sps.norm.cdf(X, loc =3, scale=0.5)
plt.plot(X,pdf)
plt.plot(X, cdf)
```
